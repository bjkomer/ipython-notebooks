## 2D Posecell Network (separated dimensions)

```python
class Clamp(object):
    """
    Output
    """
    def __init__(self, data, dt):
        self.data = data
        self.dt = dt
        L, N = data.shape
        self.output_signal = Signal( N )
        
@register_handler(Clamp)
class SimClamp(object):
    def __init__(self, nl):
        self.nl = nl
        self.t = 0.0

    def step(self, dt, J, output):
        idx = int(self.t / self.nl.dt) % len(self.nl.data)
        output[:] = self.nl.data[idx]
        self.t += dt

SINGLE_POPULATION = False        
        
class PosecellNetwork():
    
    # TODO: allow all important parameters to be specified in the init
    def __init__( self ):
        
        self.N_NEURONS = 1000
        self.N_POSITIONS = 16
        self.N_BASIS = 4
        #self.N_DIMENSIONS = 2
        self.clamp_dt = 0.001
        self.clamp_steps = 20000
        
        self.train_model()
    
    def train_model( self ):
        # 1: create signals
        self.basis_signal_x = Signal( ( self.N_BASIS * 2 ), dtype='float')
        self.basis_signal_y = Signal( ( self.N_BASIS * 2 ), dtype='float')
        self.position_signal_x = Signal(self.N_POSITIONS, dtype='float')
        self.position_signal_y = Signal(self.N_POSITIONS, dtype='float')
        self.direction_signal_x = Signal( 1, dtype='float')
        self.direction_signal_y = Signal( 1, dtype='float')
        
        # 2: create neurons
        if SINGLE_POPULATION:
            self.basis_neurons = LIF(self.N_NEURONS)
        else:
            self.basis_neurons_x = LIF(self.N_NEURONS)
            self.basis_neurons_y = LIF(self.N_NEURONS)
        
        rng = np.random.RandomState(1234)
        self.max_rates = rng.uniform(size=self.N_NEURONS, low=200, high=300)
        self.threshold = rng.uniform(size=self.N_NEURONS, low=-1.0, high=1.0)
        if SINGLE_POPULATION:
            self.basis_neurons.set_gain_bias(self.max_rates, self.threshold)
            self.basis_neurons.bias_signal.value[:] = self.basis_neurons.bias
            
             # 3: create encoder/decoder circuit
            self.enc_basis_x = Encoder(self.basis_signal_x, self.basis_neurons)
            self.enc_basis_y = Encoder(self.basis_signal_y, self.basis_neurons)
            self.enc_direc_x = Encoder(self.direction_signal_x, self.basis_neurons)
            self.enc_direc_y = Encoder(self.direction_signal_y, self.basis_neurons)
            
            self.enc_basis_x.weights *= self.basis_neurons.gain[:, None]
            self.enc_basis_y.weights *= self.basis_neurons.gain[:, None]
            self.enc_direc_x.weights *= self.basis_neurons.gain[:, None]
            self.enc_direc_y.weights *= self.basis_neurons.gain[:, None]
        else:
            self.basis_neurons_x.set_gain_bias(self.max_rates, self.threshold)
            self.basis_neurons_x.bias_signal.value[:] = self.basis_neurons_x.bias
            self.basis_neurons_y.set_gain_bias(self.max_rates, self.threshold)
            self.basis_neurons_y.bias_signal.value[:] = self.basis_neurons_y.bias
            
             # 3: create encoder/decoder circuit
            self.enc_basis_x = Encoder(self.basis_signal_x, self.basis_neurons_x)
            self.enc_basis_y = Encoder(self.basis_signal_y, self.basis_neurons_y)
            self.enc_direc_x = Encoder(self.direction_signal_x, self.basis_neurons_x)
            self.enc_direc_y = Encoder(self.direction_signal_y, self.basis_neurons_y)
            
            self.enc_basis_x.weights *= self.basis_neurons_x.gain[:, None]
            self.enc_basis_y.weights *= self.basis_neurons_y.gain[:, None]
            self.enc_direc_x.weights *= self.basis_neurons_x.gain[:, None]
            self.enc_direc_y.weights *= self.basis_neurons_y.gain[:, None]
        
       
        
        # 4: train encoder/decoder circuit
        
        # 4.1: Sample a possible trajectory
        self.data_directions_x = (rng.randn(self.clamp_steps))
        self.data_directions_x = scipy.signal.lfilter(np.ones(64)/256.0, [1], self.data_directions_x)
        self.data_positions_real_x = np.cumsum(self.data_directions_x)
        self.data_positions_table_x = np.zeros((len(self.data_positions_real_x), self.N_POSITIONS))
        for pos, table in zip(self.data_positions_real_x, self.data_positions_table_x):
            pos = pos % self.N_POSITIONS  # -- ensures non-neg, but still float
            posi = int(pos)
            posf = pos - posi
            table[posi] = 1.0 - posf
            table[(posi + 1) % self.N_POSITIONS] = posf
            
        self.data_directions_y = (rng.randn(self.clamp_steps))
        self.data_directions_y = scipy.signal.lfilter(np.ones(64)/256.0, [1], self.data_directions_y)
        self.data_positions_real_y = np.cumsum(self.data_directions_y)
        self.data_positions_table_y = np.zeros((len(self.data_positions_real_y), self.N_POSITIONS))
        for pos, table in zip(self.data_positions_real_y, self.data_positions_table_y):
            pos = pos % self.N_POSITIONS  # -- ensures non-neg, but still float
            posi = int(pos)
            posf = pos - posi
            table[posi] = 1.0 - posf
            table[(posi + 1) % self.N_POSITIONS] = posf
        
        # 4.3: Convert data positions to Fourier domain
        #blobs = scipy.signal.lfilter([.15, .7, .15], [1], data_positions_table)
        self.basis_blobs_x = scipy.fftpack.fft(self.data_positions_table_x)
        self.basis_blobs_x[:, self.N_BASIS:] = 0
        self.data_basis_x = np.empty((len(self.basis_blobs_x), 2 * self.N_BASIS))
        self.data_basis_x[:, ::2] = np.real(self.basis_blobs_x[:, :self.N_BASIS])
        self.data_basis_x[:, 1::2] = np.imag(self.basis_blobs_x[:, :self.N_BASIS])
        self.data_basis_next_x = np.zeros_like(self.data_basis_x)
        self.data_basis_next_x[:-1] = self.data_basis_x[1:]
        self.data_basis_next_x[-1] = self.data_basis_x[0]
        
        self.basis_blobs_y = scipy.fftpack.fft(self.data_positions_table_y)
        self.basis_blobs_y[:, self.N_BASIS:] = 0
        self.data_basis_y = np.empty((len(self.basis_blobs_y), 2 * self.N_BASIS))
        self.data_basis_y[:, ::2] = np.real(self.basis_blobs_y[:, :self.N_BASIS])
        self.data_basis_y[:, 1::2] = np.imag(self.basis_blobs_y[:, :self.N_BASIS])
        self.data_basis_next_y = np.zeros_like(self.data_basis_y)
        self.data_basis_next_y[:-1] = self.data_basis_y[1:]
        self.data_basis_next_y[-1] = self.data_basis_y[0]
        
        self.d2_x = scipy.fftpack.ifft(self.basis_blobs_x)
        
        self.d2_y = scipy.fftpack.ifft(self.basis_blobs_y)
        
        self.direction_coef_x = 1.0 / np.max(abs(self.data_directions_x))
        
        self.direction_coef_y = 1.0 / np.max(abs(self.data_directions_y))

        self.clamp_basis_x = Clamp(np.asarray(self.data_basis_x), dt=self.clamp_dt)
        self.clamp_direc_x = Clamp(np.asarray(self.data_directions_x)[:, None] * self.direction_coef_x,
                            dt=self.clamp_dt)
        self.clamp_basis_out_x = Clamp(np.asarray(self.data_basis_next_x), dt=self.clamp_dt)
        self.clamp_positions_x = Clamp(np.asarray(self.data_positions_table_x), dt=self.clamp_dt)
        
        self.clamp_basis_y = Clamp(np.asarray(self.data_basis_y), dt=self.clamp_dt)
        self.clamp_direc_y = Clamp(np.asarray(self.data_directions_y)[:, None] * self.direction_coef_y,
                            dt=self.clamp_dt)
        self.clamp_basis_out_y = Clamp(np.asarray(self.data_basis_next_y), dt=self.clamp_dt)
        self.clamp_positions_y = Clamp(np.asarray(self.data_positions_table_y), dt=self.clamp_dt)
        
        # 4.4: Run simulator
        training_model = SimModel()
        if SINGLE_POPULATION:
            training_model.signals.update(
                [self.basis_signal_x, self.position_signal_x, self.direction_signal_x,
                 self.basis_neurons.input_signal, self.basis_neurons.bias_signal,
                 self.basis_neurons.output_signal,
                 self.clamp_basis_x.output_signal,
                 self.clamp_direc_x.output_signal,
                 self.clamp_basis_out_x.output_signal,
                 self.clamp_positions_x.output_signal,
                 self.basis_signal_y, self.position_signal_y, self.direction_signal_y,
                 #self.basis_neurons.input_signal, self.basis_neurons.bias_signal,
                 #self.basis_neurons.output_signal,
                 self.clamp_basis_y.output_signal,
                 self.clamp_direc_y.output_signal,
                 self.clamp_basis_out_y.output_signal,
                 self.clamp_positions_y.output_signal,
                ])
            training_model.nonlinearities.update([self.basis_neurons,
                                                  self.clamp_basis_x, self.clamp_direc_x,
                                                  self.clamp_basis_out_x, self.clamp_positions_x,
                                                  self.clamp_basis_y, self.clamp_direc_y,
                                                  self.clamp_basis_out_y, self.clamp_positions_y])
        else:
            training_model.signals.update(
                [self.basis_signal_x, self.position_signal_x, self.direction_signal_x,
                 self.basis_neurons_x.input_signal, self.basis_neurons_x.bias_signal,
                 self.basis_neurons_x.output_signal,
                 self.basis_neurons_y.input_signal, self.basis_neurons_y.bias_signal,
                 self.basis_neurons_y.output_signal,
                 self.clamp_basis_x.output_signal,
                 self.clamp_direc_x.output_signal,
                 self.clamp_basis_out_x.output_signal,
                 self.clamp_positions_x.output_signal,
                 self.basis_signal_y, self.position_signal_y, self.direction_signal_y,
                 #self.basis_neurons.input_signal, self.basis_neurons.bias_signal,
                 #self.basis_neurons.output_signal,
                 self.clamp_basis_y.output_signal,
                 self.clamp_direc_y.output_signal,
                 self.clamp_basis_out_y.output_signal,
                 self.clamp_positions_y.output_signal,
                ])
            training_model.nonlinearities.update([self.basis_neurons_x, self.basis_neurons_y,
                                                  self.clamp_basis_x, self.clamp_direc_x,
                                                  self.clamp_basis_out_x, self.clamp_positions_x,
                                                  self.clamp_basis_y, self.clamp_direc_y,
                                                  self.clamp_basis_out_y, self.clamp_positions_y])
            
        training_model.encoders.update([self.enc_basis_x, self.enc_direc_x,
                                        self.enc_basis_y, self.enc_direc_y])
        training_model.decoder(self.clamp_basis_x, self.basis_signal_x, 1.0)
        training_model.decoder(self.clamp_direc_x, self.direction_signal_x, 1.0)
        training_model.transform(1.0, self.direction_signal_x, self.direction_signal_x)
        
        training_model.decoder(self.clamp_basis_y, self.basis_signal_y, 1.0)
        training_model.decoder(self.clamp_direc_y, self.direction_signal_y, 1.0)
        training_model.transform(1.0, self.direction_signal_y, self.direction_signal_y)
        
        
        simple_filter_decoded_signal(training_model, self.basis_signal_x, pstc=.01)
        simple_filter_decoded_signal(training_model, self.position_signal_x, pstc=.01)
        
        simple_filter_decoded_signal(training_model, self.basis_signal_y, pstc=.01)
        simple_filter_decoded_signal(training_model, self.position_signal_y, pstc=.01)
        
        p_clamp_direc = probe_nl(training_model, self.clamp_direc_x)
        p_clamp_basis = probe_nl(training_model, self.clamp_basis_x)
        p_basis_neurons = None
        if SINGLE_POPULATION:
            p_basis_neurons = probe_nl(training_model, self.basis_neurons)
        else:
            p_basis_neurons = []
            p_basis_neurons.append( probe_nl(training_model, self.basis_neurons_x) )
            p_basis_neurons.append( probe_nl(training_model, self.basis_neurons_y) )
        p_basis_signal = training_model.probe(self.basis_signal_x, dt=0)
        p_clamp_basis_out_x = probe_nl(training_model, self.clamp_basis_out_x)
        p_clamp_positions_x = probe_nl(training_model, self.clamp_positions_x)
        
        p_clamp_basis_out_y = probe_nl(training_model, self.clamp_basis_out_y)
        p_clamp_positions_y = probe_nl(training_model, self.clamp_positions_y)
        
        sim = Simulator(training_model)
        sim.run_steps(self.clamp_steps)
        
        if SINGLE_POPULATION:
        
            A = np.asarray(sim.probe_data(p_basis_neurons))
            d_clamp_basis_out_x = sim.probe_data(p_clamp_basis_out_x)
            d_clamp_positions_x = sim.probe_data(p_clamp_positions_x)
            
            d_clamp_basis_out_y = sim.probe_data(p_clamp_basis_out_y)
            d_clamp_positions_y = sim.probe_data(p_clamp_positions_y)
            
            basis_readout_weights_x, res5, rank5, s5 = np.linalg.lstsq(A,
                                                                       d_clamp_basis_out_x,
                                                                       rcond=0.01)
            posit_readout_weights_x, res6, rank6, s6 = np.linalg.lstsq(A,
                                                                       d_clamp_positions_x,
                                                                       rcond=0.01)
    
            basis_readout_weights_y, res5, rank5, s5 = np.linalg.lstsq(A,
                                                                       d_clamp_basis_out_y,
                                                                       rcond=0.01)
            posit_readout_weights_y, res6, rank6, s6 = np.linalg.lstsq(A,
                                                                       d_clamp_positions_y,
                                                                       rcond=0.01)
    
            self.dec_basis_x = Decoder(self.basis_neurons, self.basis_signal_x,
                                     weights=basis_readout_weights_x.T)
            self.dec_posit_x = Decoder(self.basis_neurons, self.position_signal_x,
                                     weights=posit_readout_weights_x.T)
            
            self.dec_basis_y = Decoder(self.basis_neurons, self.basis_signal_y,
                                     weights=basis_readout_weights_y.T)
            self.dec_posit_y = Decoder(self.basis_neurons, self.position_signal_y,
                                     weights=posit_readout_weights_y.T)
            
        else:
        
            A_x = np.asarray(sim.probe_data(p_basis_neurons[0]))
            A_y = np.asarray(sim.probe_data(p_basis_neurons[1]))
            d_clamp_basis_out_x = sim.probe_data(p_clamp_basis_out_x)
            d_clamp_positions_x = sim.probe_data(p_clamp_positions_x)
            
            d_clamp_basis_out_y = sim.probe_data(p_clamp_basis_out_y)
            d_clamp_positions_y = sim.probe_data(p_clamp_positions_y)
            
            basis_readout_weights_x, res5, rank5, s5 = np.linalg.lstsq(A_x,
                                                                       d_clamp_basis_out_x,
                                                                       rcond=0.01)
            posit_readout_weights_x, res6, rank6, s6 = np.linalg.lstsq(A_x,
                                                                       d_clamp_positions_x,
                                                                       rcond=0.01)
    
            basis_readout_weights_y, res5, rank5, s5 = np.linalg.lstsq(A_y,
                                                                       d_clamp_basis_out_y,
                                                                       rcond=0.01)
            posit_readout_weights_y, res6, rank6, s6 = np.linalg.lstsq(A_y,
                                                                       d_clamp_positions_y,
                                                                       rcond=0.01)
    
            self.dec_basis_x = Decoder(self.basis_neurons_x, self.basis_signal_x,
                                     weights=basis_readout_weights_x.T)
            self.dec_posit_x = Decoder(self.basis_neurons_x, self.position_signal_x,
                                     weights=posit_readout_weights_x.T)
            
            self.dec_basis_y = Decoder(self.basis_neurons_y, self.basis_signal_y,
                                     weights=basis_readout_weights_y.T)
            self.dec_posit_y = Decoder(self.basis_neurons_y, self.position_signal_y,
                                     weights=posit_readout_weights_y.T)
    
    def test_model( self, direction_data, injected_data=None, steps=1000, b_ratio=0.97 ):
        """ direction_data is a numpy array
            injected_data is a tuple of numpy arrays """
        
        data_injected_fourier_pos_x = np.zeros( ( self.clamp_steps, 2 * self.N_BASIS ) )
        data_injected_fourier_pos_y = np.zeros( ( self.clamp_steps, 2 * self.N_BASIS ) )
        if injected_data != None:
            data_injected_fourier_pos_x = build_table_2d( injected_data[0] )
            data_injected_fourier_pos_y = build_table_2d( injected_data[1] )

        self.clamp_injected_fourier_pos_x = Clamp(np.asarray(data_injected_fourier_pos_x), dt=self.clamp_dt)
        self.clamp_injected_fourier_pos_y = Clamp(np.asarray(data_injected_fourier_pos_y), dt=self.clamp_dt)
        
        self.testing_model = SimModel()
        
        if SINGLE_POPULATION:
        
            self.testing_model.signals.update(
                [self.basis_signal_x, self.position_signal_x, self.direction_signal_x,
                 self.basis_neurons.input_signal, self.basis_neurons.bias_signal,
                 self.basis_neurons.output_signal, self.clamp_injected_fourier_pos_x.output_signal,
                 self.basis_signal_y, self.position_signal_y, self.direction_signal_y,
                 self.clamp_injected_fourier_pos_y.output_signal
                ])
            self.testing_model.nonlinearities.update([self.basis_neurons, self.clamp_injected_fourier_pos_x, 
                                                      self.clamp_injected_fourier_pos_y])
            
        else:
        
            self.testing_model.signals.update(
                [self.basis_signal_x, self.position_signal_x, self.direction_signal_x,
                 self.basis_neurons_x.input_signal, self.basis_neurons_x.bias_signal,
                 self.basis_neurons_x.output_signal,
                 self.basis_neurons_y.input_signal, self.basis_neurons_y.bias_signal,
                 self.basis_neurons_y.output_signal, self.clamp_injected_fourier_pos_x.output_signal,
                 self.basis_signal_y, self.position_signal_y, self.direction_signal_y,
                 self.clamp_injected_fourier_pos_y.output_signal
                ])
            self.testing_model.nonlinearities.update([self.basis_neurons_x, self.basis_neurons_y, 
                                                      self.clamp_injected_fourier_pos_x, 
                                                      self.clamp_injected_fourier_pos_y])
            
        self.testing_model.encoders.update([self.enc_basis_x, self.enc_direc_x, self.enc_basis_y, self.enc_direc_y])
        self.testing_model.decoders.update([self.dec_basis_x, self.dec_posit_x, self.dec_basis_y, self.dec_posit_y])
        
        simple_filter_decoded_signal(self.testing_model, self.position_signal_x, pstc=.005)
        simple_filter_decoded_signal(self.testing_model, self.position_signal_y, pstc=.005)
        
        # -- add direction clamp (using feedback for basis)
        self.clamp_direc_x.data = direction_data[0]
        self.clamp_direc_y.data = direction_data[1]
        self.testing_model.signals.add(self.clamp_direc_x.output_signal)
        self.testing_model.nonlinearities.add(self.clamp_direc_x)
        self.testing_model.decoder(self.clamp_direc_x, self.direction_signal_x, 1.0)
        self.testing_model.decoder(self.clamp_injected_fourier_pos_x, self.clamp_injected_fourier_pos_x.output_signal, 1.0)
        
        self.testing_model.signals.add(self.clamp_direc_y.output_signal)
        self.testing_model.nonlinearities.add(self.clamp_direc_y)
        self.testing_model.decoder(self.clamp_direc_y, self.direction_signal_y, 1.0)
        self.testing_model.decoder(self.clamp_injected_fourier_pos_y, self.clamp_injected_fourier_pos_y.output_signal, 1.0)
        
        self.testing_model.transform(1.0, self.direction_signal_x, self.direction_signal_x)
        a, b = filter_coefs(0.005, self.testing_model.dt)
        self.testing_model.filter(a, self.basis_signal_x, self.basis_signal_x)
        self.testing_model.transform(b*b_ratio, self.basis_signal_x, self.basis_signal_x)
        self.testing_model.transform(b*(1-b_ratio), self.clamp_injected_fourier_pos_x.output_signal, self.basis_signal_x)
        
        self.testing_model.transform(1.0, self.direction_signal_y, self.direction_signal_y)
        self.testing_model.filter(a, self.basis_signal_y, self.basis_signal_y)
        self.testing_model.transform(b*b_ratio, self.basis_signal_y, self.basis_signal_y)
        self.testing_model.transform(b*(1-b_ratio), self.clamp_injected_fourier_pos_y.output_signal, self.basis_signal_y)
        
        self.tp_basis_x = self.testing_model.probe(self.basis_signal_x, dt=0)
        self.tp_posit_x = self.testing_model.probe(self.position_signal_x, dt=0)
        if SINGLE_POPULATION:
            self.tp_neurons = probe_nl(self.testing_model, self.basis_neurons)
        else:
            self.tp_neurons_x = probe_nl(self.testing_model, self.basis_neurons_x)
            self.tp_neurons_y = probe_nl(self.testing_model, self.basis_neurons_y)
        self.tp_dir_x = self.testing_model.probe(self.direction_signal_x, dt=0)
        
        self.tp_basis_y = self.testing_model.probe(self.basis_signal_y, dt=0)
        self.tp_posit_y = self.testing_model.probe(self.position_signal_y, dt=0)
        self.tp_dir_y = self.testing_model.probe(self.direction_signal_y, dt=0)
        
        self.test_sim = Simulator(self.testing_model)
        self.test_sim.run_steps( steps )
        
        self.calculated_position_x = get_position( self.test_sim.probe_data(self.tp_posit_x).T, precise=True )
        self.theoretical_position_x = (np.cumsum(self.test_sim.probe_data(self.tp_dir_x).ravel()) *(32.0/256.0)) % self.N_POSITIONS
        
        self.calculated_position_y = get_position( self.test_sim.probe_data(self.tp_posit_y).T, precise=True )
        self.theoretical_position_y = (np.cumsum(self.test_sim.probe_data(self.tp_dir_y).ravel()) *(32.0/256.0)) % self.N_POSITIONS
    
    def discrepancy( self, ratio, direction_data, injected_data, steps=1000 ):
        """ Runs the test model, using 'ratio' as the ratio of the 'b' value of the basis_signal and 
            clamp_injected_fourier_pos signal transform 
            This function returns a value representing how close the model results are to what the 'actual'
            result should be.
            This function is designed to be used by hyperopt """
        
        self.test_model( direction_data, injected_data, steps=steps, b_ratio=ratio )
        
        # Check at different offsets of theoretical position, since the start point is random
        min_val = float("inf")
        density = 2
        for i in xrange( self.N_POSITIONS * density ):
            # TODO: might need a third option of subtracting in the np.min() statement
            diff = np.minimum( np.abs( self.calculated_position - \
                                      ( self.theoretical_position + i / float( density ) ) % self.N_POSITIONS ),  
                               np.abs( self.calculated_position + self.N_POSITIONS - \
                                      ( self.theoretical_position + i / float( density ) ) % self.N_POSITIONS ) )
            val = np.sum( diff )
            if val < min_val:
                min_val = val
                
        return min_val

    def hyperopt_wrapper( self, ratio ):
        
        direction_data = np.ones( self.clamp_steps ) * 0.3
        
        # Test with no injected data
        injected_data = None
        
        return self.discrepancy( ratio, direction_data, injected_data, steps=1000 )
    
    def plot_results_x( self ):
        
        fig = plt.figure(figsize=(30,15), dpi=1000)
        plt.subplot(2, 4, 1)
        plt.plot(self.data_positions_real_x)
        plt.title('pos(t)')
        plt.subplot(2, 4, 2)
        plt.imshow(self.data_basis_x, aspect='auto')
        plt.title('fft(pos table)')
        plt.subplot(2, 4, 3)
        plt.imshow(np.real(self.d2_x).T, origin='lower', aspect='auto')
        plt.title('ifft(fft(pos table))')
    
        plt.subplot(2, 4, 5)
        plt.plot(self.test_sim.probe_data(self.tp_dir_x).ravel())
        plt.title('v(t)')
    
        plt.subplot(2, 4, 6)
        plt.imshow(self.test_sim.probe_data(self.tp_basis_x), aspect='auto')
        plt.title('basis readout')
    
        plt.subplot(2, 4, 7)
        plt.imshow(self.test_sim.probe_data(self.tp_posit_x).T, origin='lower', aspect='auto')
        plt.title('position readout')
        
        plt.subplot(2, 4, 8)
        plt.plot( self.calculated_position_x )
        plt.title('calculated position')
        
        plt.subplot(2, 4, 4)
        #plt.plot( (np.cumsum(self.test_sim.probe_data(self.tp_dir).ravel()) *(64.0/256.0)) % self.N_POSITIONS )
        plt.plot( self.theoretical_position_x )
        plt.title('theoretical position')
    
        plt.show()
        
    def plot_results_y( self ):
        
        fig = plt.figure(figsize=(30,15), dpi=1000)
        plt.subplot(2, 4, 1)
        plt.plot(self.data_positions_real_y)
        plt.title('pos(t)')
        plt.subplot(2, 4, 2)
        plt.imshow(self.data_basis_y, aspect='auto')
        plt.title('fft(pos table)')
        plt.subplot(2, 4, 3)
        plt.imshow(np.real(self.d2_y).T, origin='lower', aspect='auto')
        plt.title('ifft(fft(pos table))')
    
        plt.subplot(2, 4, 5)
        plt.plot(self.test_sim.probe_data(self.tp_dir_y).ravel())
        plt.title('v(t)')
    
        plt.subplot(2, 4, 6)
        plt.imshow(self.test_sim.probe_data(self.tp_basis_y), aspect='auto')
        plt.title('basis readout')
    
        plt.subplot(2, 4, 7)
        plt.imshow(self.test_sim.probe_data(self.tp_posit_y).T, origin='lower', aspect='auto')
        plt.title('position readout')
        
        plt.subplot(2, 4, 8)
        plt.plot( self.calculated_position_y )
        plt.title('calculated position')
        
        plt.subplot(2, 4, 4)
        #plt.plot( (np.cumsum(self.test_sim.probe_data(self.tp_dir).ravel()) *(64.0/256.0)) % self.N_POSITIONS )
        plt.plot( self.theoretical_position_y )
        plt.title('theoretical position')
    
        plt.show()
        
    def plot_results_2d( self ):
        
        fig = plt.figure(figsize=(30,15), dpi=1000)
        plt.subplot(2, 4, 1)
        plt.plot(self.data_positions_real)
        plt.title('pos(t)')
        plt.subplot(2, 4, 2)
        plt.imshow(self.data_basis[:,:,0], aspect='auto')
        plt.title('fft(pos table)')
        plt.subplot(2, 4, 3)
        plt.imshow(np.real(self.d2)[:,:,0].T, origin='lower', aspect='auto')
        plt.title('ifft(fft(pos table))')
    
        plt.subplot(2, 4, 5)
        plt.plot(self.test_sim.probe_data(self.tp_dir).ravel())
        plt.title('v(t)')
    
        plt.subplot(2, 4, 6)
        plt.imshow(self.test_sim.probe_data(self.tp_basis), aspect='auto')
        plt.title('basis readout')
    
        plt.subplot(2, 4, 7)
        plt.imshow(self.test_sim.probe_data(self.tp_posit).T, origin='lower', aspect='auto')
        plt.title('position readout')
        
        plt.subplot(2, 4, 8)
        plt.plot( self.calculated_position )
        plt.title('calculated position')
        
        plt.subplot(2, 4, 4)
        #plt.plot( (np.cumsum(self.test_sim.probe_data(self.tp_dir).ravel()) *(64.0/256.0)) % self.N_POSITIONS )
        plt.plot( self.theoretical_position )
        plt.title('theoretical position')
    
        plt.show()
        
    def plot_2d_from_split( self, t ):
        
        x = self.test_sim.probe_data(self.tp_posit_x)[t,:]
        y = self.test_sim.probe_data(self.tp_posit_y)[t,:]
        pos = numpy.empty((len(x), len(y)))
        
        for i in xrange(len(x)):
            for j in xrange(len(y)):
                pos[i,j] = x[i] * y[j]
        
        fig = plt.figure(figsize=(20,10), dpi=1000)
        
        plt.subplot(2, 4, 7)
        plt.imshow(pos, origin='lower', aspect='auto')
        plt.title('position readout')
        
        plt.show()
        
        return pos
        
    def video_2d_from_split( self, frames=200, interval=20, step=5 ):
        
        x = self.test_sim.probe_data(self.tp_posit_x)[30,:]
        y = self.test_sim.probe_data(self.tp_posit_y)[30,:]
        init_pos = numpy.empty((len(x), len(y)))
            
        for i in xrange(len(x)):
            for j in xrange(len(y)):
                init_pos[i,j] = x[i] * y[j]       
        
        fig = plt.figure(figsize=(20,10), dpi=1000)
        ax = plt.subplot(111)
        #NOTE: if initalized to 0, the animation will break for some reason. Initializing to random as a workaround
        im = ax.imshow( init_pos )
        #im = ax.imshow( np.random.randint(10, size=(self.N_POSITIONS, self.N_POSITIONS) ) )
        
        def init():
            im.set_data( init_pos )
            #im.set_data( np.random.randint(10, size=(self.N_POSITIONS, self.N_POSITIONS) ) )
            return im
        
        def animate( t ):
            x = self.test_sim.probe_data(self.tp_posit_x)[t*step,:]
            y = self.test_sim.probe_data(self.tp_posit_y)[t*step,:]
            #x = self.test_sim.probe_data(self.tp_posit_x)[30,:]
            #y = self.test_sim.probe_data(self.tp_posit_y)[30,:]
            pos = numpy.empty((len(x), len(y)))
            
            for i in xrange(len(x)):
                for j in xrange(len(y)):
                    pos[i,j] = x[i] * y[j]
                    
            im.set_data( pos )
            return im
        
        anim = animation.FuncAnimation(fig, animate, init_func=init, frames=frames, interval=interval, blit=True)
        
        return anim
        #display_animation( anim )
        
        #plt.show()
            
        
posecells = PosecellNetwork()
```

```python
class Clamp(object):
    """
    Output
    """
    def __init__(self, data, dt):
        self.data = data
        self.dt = dt
        #L, N, D = data.shape
        #self.output_signal = Signal( N * D )
        print(data.shape)
        L, N = data.shape
        self.output_signal = Signal( N )
        
@register_handler(Clamp)
class SimClamp(object):
    def __init__(self, nl):
        self.nl = nl
        self.t = 0.0

    def step(self, dt, J, output):
        idx = int(self.t / self.nl.dt) % len(self.nl.data)
        output[:] = self.nl.data[idx]
        self.t += dt
        #for i in xrange( self.nl.data.shape[1] ):
        #    for j in xrange( self.nl.data.shape[1] ):
        #        #print( self.nl.data[idx,i,0] * self.nl.data[idx,j,1] )
        #        output[ i + j * self.nl.data.shape[1] ] = self.nl.data[idx,i,0] * self.nl.data[idx,j,1]
        
        #s = len(self.nl.data[idx,:,0])
        #output[:s] = self.nl.data[idx,:,0]
        #output[s:] = self.nl.data[idx,:,1]

class PosecellNetwork():
    
    # TODO: allow all important parameters to be specified in the init
    def __init__( self ):
        
        self.N_NEURONS = 1000
        self.N_POSITIONS = 16
        self.N_BASIS = 4
        self.N_DIMENSIONS = 2
        self.clamp_dt = 0.001
        self.clamp_steps = 20000
        
        self.train_model()
    
    def train_model( self ):
        # 1: create signals
        self.basis_signal = Signal( ( self.N_BASIS * 2 ) * self.N_DIMENSIONS, dtype='float')
        self.position_signal = Signal(self.N_POSITIONS * self.N_DIMENSIONS, dtype='float')
        self.direction_signal = Signal( 1 * self.N_DIMENSIONS, dtype='float')
        
        # 2: create neurons
        self.basis_neurons = LIF(self.N_NEURONS)
        
        rng = np.random.RandomState(1234)
        self.max_rates = rng.uniform(size=self.N_NEURONS, low=200, high=300)
        self.threshold = rng.uniform(size=self.N_NEURONS, low=-1.0, high=1.0)
        self.basis_neurons.set_gain_bias(self.max_rates, self.threshold)
        self.basis_neurons.bias_signal.value[:] = self.basis_neurons.bias
        
        # 3: create encoder/decoder circuit
        self.enc_basis = Encoder(self.basis_signal, self.basis_neurons)
        self.enc_direc = Encoder(self.direction_signal, self.basis_neurons)
        
        self.enc_basis.weights *= self.basis_neurons.gain[:, None]
        self.enc_direc.weights *= self.basis_neurons.gain[:, None]
        
        # 4: train encoder/decoder circuit
        
        # 4.1: Sample a possible trajectory
        self.data_directions = (rng.randn(self.clamp_steps, self.N_DIMENSIONS))
        self.data_directions = scipy.signal.lfilter(np.ones(64)/256.0, [1], self.data_directions)
        self.data_positions_real = np.cumsum(self.data_directions, axis=0)
        #self.data_positions_table = np.zeros((len(self.data_positions_real), self.N_POSITIONS, self.N_DIMENSIONS))
        """
        self.data_positions_table = np.zeros((len(self.data_positions_real), self.N_POSITIONS, self.N_POSITIONS))
        for pos, table in zip(self.data_positions_real, self.data_positions_table):
            pos_x = pos[0]
            pos_y = pos[1]
            pos_x = pos_x % self.N_POSITIONS  # -- ensures non-neg, but still float
            pos_y = pos_y % self.N_POSITIONS  # -- ensures non-neg, but still float
            posi_x = int(pos_x)
            posi_y = int(pos_y)
            posf_x = pos_x - posi_x
            posf_y = pos_y - posi_y
            table[posi_x, posi_y ] = ( 2 - posf_x - posf_y ) / 2
            table[(posi_x + 1) % self.N_POSITIONS, (posi_y + 1) % self.N_POSITIONS] = ( posf_x + posf_y ) / 2
            table[posi_x, (posi_y + 1) % self.N_POSITIONS] = ( 1 - posf_x + posf_y ) / 2
            table[(posi_x + 1) % self.N_POSITIONS, posi_y] = ( posf_x + 1 - posf_y ) / 2
        """
        self.data_positions_table = np.zeros((len(self.data_positions_real), self.N_POSITIONS, self.N_DIMENSIONS))
        for pos, table in zip(self.data_positions_real, self.data_positions_table):
            pos_x = pos[0]
            pos_y = pos[1]
            pos_x = pos_x % self.N_POSITIONS  # -- ensures non-neg, but still float
            pos_y = pos_y % self.N_POSITIONS  # -- ensures non-neg, but still float
            posi_x = int(pos_x)
            posi_y = int(pos_y)
            posf_x = pos_x - posi_x
            posf_y = pos_y - posi_y
            table[posi_x, 0 ] = 1 - posf_x
            table[posi_y, 1 ] = 1 - posf_y
            table[(posi_x + 1) % self.N_POSITIONS, 0 ] = posf_x
            table[(posi_y + 1) % self.N_POSITIONS, 1 ] = posf_y

        # 4.3: Convert data positions to Fourier domain
        #blobs = scipy.signal.lfilter([.15, .7, .15], [1], data_positions_table)
        ##self.basis_blobs = scipy.fftpack.fft(self.data_positions_table)
        self.basis_blobs = scipy.fftpack.fft(self.data_positions_table[:,:,0])
        self.basis_blobs[:, self.N_BASIS:] = 0
        ##self.data_basis = np.empty((len(self.basis_blobs), 2 * self.N_BASIS, self.N_DIMENSIONS))
        self.data_basis = np.empty((len(self.basis_blobs), 2 * self.N_BASIS))
        #self.data_basis = np.empty((len(self.basis_blobs), 2 * self.N_BASIS, self.N_POSITIONS))
        self.data_basis[:, ::2] = np.real(self.basis_blobs[:, :self.N_BASIS])
        self.data_basis[:, 1::2] = np.imag(self.basis_blobs[:, :self.N_BASIS])
        self.data_basis_next = np.zeros_like(self.data_basis)
        self.data_basis_next[:-1] = self.data_basis[1:]
        self.data_basis_next[-1] = self.data_basis[0]
        
        self.d2 = scipy.fftpack.ifft(self.basis_blobs)
        
        self.direction_coef = 1.0 / np.max(abs(self.data_directions))
        
        self.clamp_basis = Clamp(np.asarray(self.data_basis), dt=self.clamp_dt)
        #self.clamp_direc = Clamp(np.asarray(self.data_directions)[:, None] * self.direction_coef,
        #                    dt=self.clamp_dt)
        self.clamp_direc = Clamp(np.asarray(self.data_directions) * self.direction_coef,
                            dt=self.clamp_dt)
        self.clamp_basis_out = Clamp(np.asarray(self.data_basis_next), dt=self.clamp_dt)
        # TEMP, for now ignore the y dimension
        #self.clamp_positions = Clamp(np.asarray(self.data_positions_table), dt=self.clamp_dt)
        self.clamp_positions = Clamp(np.asarray(self.data_positions_table[:,:,0]), dt=self.clamp_dt)
        
        # 4.4: Run simulator
        training_model = SimModel()
        #nl_readout = training_model.signal(N_NEURONS)
        training_model.signals.update(
            [self.basis_signal, self.position_signal, self.direction_signal,
             self.basis_neurons.input_signal, self.basis_neurons.bias_signal,
             self.basis_neurons.output_signal,
             self.clamp_basis.output_signal,
             self.clamp_direc.output_signal,
             self.clamp_basis_out.output_signal,
             self.clamp_positions.output_signal,
            ])
        training_model.nonlinearities.update([self.basis_neurons,
                                              self.clamp_basis, self.clamp_direc,
                                              self.clamp_basis_out, self.clamp_positions])
        training_model.encoders.update([self.enc_basis, self.enc_direc])
        training_model.decoder(self.clamp_basis, self.basis_signal, 1.0)
        training_model.decoder(self.clamp_direc, self.direction_signal, 1.0)
        training_model.transform(1.0, self.direction_signal, self.direction_signal)
        #training_model.decoder(basis_neurons, nl_readout, 1.0)
        #training_model.transform(1.0, nl_readout, nl_readout)
        
        simple_filter_decoded_signal(training_model, self.basis_signal, pstc=.01)
        simple_filter_decoded_signal(training_model, self.position_signal, pstc=.01)
        
        p_clamp_direc = probe_nl(training_model, self.clamp_direc)
        p_clamp_basis = probe_nl(training_model, self.clamp_basis)
        p_basis_neurons = probe_nl(training_model, self.basis_neurons)
        p_basis_signal = training_model.probe(self.basis_signal, dt=0)
        p_clamp_basis_out = probe_nl(training_model, self.clamp_basis_out)
        p_clamp_positions = probe_nl(training_model, self.clamp_positions)
        
        sim = Simulator(training_model)
        sim.run_steps(self.clamp_steps)
        
        A = np.asarray(sim.probe_data(p_basis_neurons))
        d_clamp_basis_out = sim.probe_data(p_clamp_basis_out)
        d_clamp_positions = sim.probe_data(p_clamp_positions)
        
        basis_readout_weights, res5, rank5, s5 = np.linalg.lstsq(A,
                                                                 d_clamp_basis_out,
                                                                 rcond=0.01)
        posit_readout_weights, res6, rank6, s6 = np.linalg.lstsq(A,
                                                                 d_clamp_positions,
                                                                 rcond=0.01)
        
        self.dec_basis = Decoder(self.basis_neurons, self.basis_signal,
                                 weights=basis_readout_weights.T)
        self.dec_posit = Decoder(self.basis_neurons, self.position_signal,
                                 weights=posit_readout_weights.T)
    
    def test_model( self, direction_data, injected_data=None, steps=1000, b_ratio=0.97 ):
        """ direction_data is a numpy array
            injected_data is a tuple of numpy arrays """
        
        data_injected_fourier_pos = np.zeros( ( self.clamp_steps, 2 * self.N_BASIS, self.N_DIMENSIONS ) )
        if injected_data != None:
            data_injected_fourier_pos = build_table_2d( injected_data )

        self.clamp_injected_fourier_pos = Clamp(np.asarray(data_injected_fourier_pos), dt=self.clamp_dt)
        
        self.testing_model = SimModel()
        self.testing_model.signals.update(
            [self.basis_signal, self.position_signal, self.direction_signal,
             self.basis_neurons.input_signal, self.basis_neurons.bias_signal,
             self.basis_neurons.output_signal, self.clamp_injected_fourier_pos.output_signal
            ])
        self.testing_model.nonlinearities.update([self.basis_neurons, self.clamp_injected_fourier_pos])
        self.testing_model.encoders.update([self.enc_basis, self.enc_direc])
        self.testing_model.decoders.update([self.dec_basis, self.dec_posit])
        
        simple_filter_decoded_signal(self.testing_model, self.position_signal, pstc=.005)
        
        # -- add direction clamp (using feedback for basis)
        self.clamp_direc.data = direction_data
        self.testing_model.signals.add(self.clamp_direc.output_signal)
        self.testing_model.nonlinearities.add(self.clamp_direc)
        self.testing_model.decoder(self.clamp_direc, self.direction_signal, 1.0)
        self.testing_model.decoder(self.clamp_injected_fourier_pos, self.clamp_injected_fourier_pos.output_signal, 1.0)
        
        self.testing_model.transform(1.0, self.direction_signal, self.direction_signal)
        a, b = filter_coefs(0.005, self.testing_model.dt)
        self.testing_model.filter(a, self.basis_signal, self.basis_signal)
        self.testing_model.transform(b*b_ratio, self.basis_signal, self.basis_signal)
        self.testing_model.transform(b*(1-b_ratio), self.clamp_injected_fourier_pos.output_signal, self.basis_signal)
        
        self.tp_basis = self.testing_model.probe(self.basis_signal, dt=0)
        self.tp_posit = self.testing_model.probe(self.position_signal, dt=0)
        self.tp_neurons = probe_nl(self.testing_model, self.basis_neurons)
        self.tp_dir = self.testing_model.probe(self.direction_signal, dt=0)
        
        self.test_sim = Simulator(self.testing_model)
        self.test_sim.run_steps( steps )
        
        self.calculated_position = get_position_2d( self.test_sim.probe_data(self.tp_posit).T, precise=True )
        self.theoretical_position = (np.cumsum(self.test_sim.probe_data(self.tp_dir).ravel()) *(32.0/256.0)) % self.N_POSITIONS
    
    def discrepancy( self, ratio, direction_data, injected_data, steps=1000 ):
        """ Runs the test model, using 'ratio' as the ratio of the 'b' value of the basis_signal and 
            clamp_injected_fourier_pos signal transform 
            This function returns a value representing how close the model results are to what the 'actual'
            result should be.
            This function is designed to be used by hyperopt """
        
        self.test_model( direction_data, injected_data, steps=steps, b_ratio=ratio )
        
        # Check at different offsets of theoretical position, since the start point is random
        min_val = float("inf")
        density = 2
        for i in xrange( self.N_POSITIONS * density ):
            # TODO: might need a third option of subtracting in the np.min() statement
            diff = np.minimum( np.abs( self.calculated_position - \
                                      ( self.theoretical_position + i / float( density ) ) % self.N_POSITIONS ),  
                               np.abs( self.calculated_position + self.N_POSITIONS - \
                                      ( self.theoretical_position + i / float( density ) ) % self.N_POSITIONS ) )
            val = np.sum( diff )
            if val < min_val:
                min_val = val
                
        return min_val

    def hyperopt_wrapper( self, ratio ):
        
        direction_data = np.ones( self.clamp_steps ) * 0.3
        
        # Test with no injected data
        injected_data = None
        
        return self.discrepancy( ratio, direction_data, injected_data, steps=1000 )
    
    def plot_results( self ):
        
        fig = plt.figure(figsize=(30,15), dpi=1000)
        plt.subplot(2, 4, 1)
        plt.plot(self.data_positions_real)
        plt.title('pos(t)')
        plt.subplot(2, 4, 2)
        plt.imshow(self.data_basis, aspect='auto')
        plt.title('fft(pos table)')
        plt.subplot(2, 4, 3)
        plt.imshow(np.real(self.d2).T, origin='lower', aspect='auto')
        plt.title('ifft(fft(pos table))')
    
        plt.subplot(2, 4, 5)
        plt.plot(self.test_sim.probe_data(self.tp_dir).ravel())
        plt.title('v(t)')
    
        plt.subplot(2, 4, 6)
        plt.imshow(self.test_sim.probe_data(self.tp_basis), aspect='auto')
        plt.title('basis readout')
    
        plt.subplot(2, 4, 7)
        plt.imshow(self.test_sim.probe_data(self.tp_posit).T, origin='lower', aspect='auto')
        plt.title('position readout')
        
        plt.subplot(2, 4, 8)
        plt.plot( self.calculated_position )
        plt.title('calculated position')
        
        plt.subplot(2, 4, 4)
        #plt.plot( (np.cumsum(self.test_sim.probe_data(self.tp_dir).ravel()) *(64.0/256.0)) % self.N_POSITIONS )
        plt.plot( self.theoretical_position )
        plt.title('theoretical position')
    
        plt.show()
        
    def plot_results_2d( self ):
        
        fig = plt.figure(figsize=(30,15), dpi=1000)
        plt.subplot(2, 4, 1)
        plt.plot(self.data_positions_real)
        plt.title('pos(t)')
        plt.subplot(2, 4, 2)
        plt.imshow(self.data_basis[:,:,0], aspect='auto')
        plt.title('fft(pos table)')
        plt.subplot(2, 4, 3)
        plt.imshow(np.real(self.d2)[:,:,0].T, origin='lower', aspect='auto')
        plt.title('ifft(fft(pos table))')
    
        plt.subplot(2, 4, 5)
        plt.plot(self.test_sim.probe_data(self.tp_dir).ravel())
        plt.title('v(t)')
    
        plt.subplot(2, 4, 6)
        plt.imshow(self.test_sim.probe_data(self.tp_basis), aspect='auto')
        plt.title('basis readout')
    
        plt.subplot(2, 4, 7)
        plt.imshow(self.test_sim.probe_data(self.tp_posit).T, origin='lower', aspect='auto')
        plt.title('position readout')
        
        plt.subplot(2, 4, 8)
        plt.plot( self.calculated_position )
        plt.title('calculated position')
        
        plt.subplot(2, 4, 4)
        #plt.plot( (np.cumsum(self.test_sim.probe_data(self.tp_dir).ravel()) *(64.0/256.0)) % self.N_POSITIONS )
        plt.plot( self.theoretical_position )
        plt.title('theoretical position')
    
        plt.show()
        
    def sandbox_plots( self ):
        
        fig = plt.figure(figsize=(30,15), dpi=1000)
        
        plt.subplot(2, 4, 2)
        plt.imshow(self.data_basis[:,0,:], aspect='auto')
        plt.title('fft(pos table)')
        plt.subplot(2, 4, 3)
        plt.imshow(np.real(self.d2)[:,0,:].T, origin='lower', aspect='auto')
        plt.title('ifft(fft(pos table))')
        
        plt.subplot(2, 4, 6)
        plt.imshow(self.data_basis[0,:,:], aspect='auto')
        plt.title('fft(pos table)')
        plt.subplot(2, 4, 7)
        plt.imshow(np.real(self.d2)[0,:,:].T, origin='lower', aspect='auto')
        plt.title('ifft(fft(pos table))')
        
        plt.show()
        
posecells = PosecellNetwork()
```

```python
class PosecellNetwork():
    
    # TODO: allow all important parameters to be specified in the init
    def __init__( self ):
        
        self.N_NEURONS = 1000
        self.N_POSITIONS = 16
        self.N_BASIS = 4
        self.N_DIMENSIONS = 2
        self.clamp_dt = 0.001
        self.clamp_steps = 20000
        
        self.train_model()
    
    def train_model( self ):
        # 1: create signals
        self.basis_signal = Signal( ( self.N_BASIS * 2 ) ** self.N_DIMENSIONS, dtype='float')
        self.position_signal = Signal(self.N_POSITIONS ** self.N_DIMENSIONS, dtype='float')
        self.direction_signal = Signal( 1 ** self.N_DIMENSIONS, dtype='float')
        
        # 2: create neurons
        self.basis_neurons = LIF(self.N_NEURONS)
        
        rng = np.random.RandomState(1234)
        self.max_rates = rng.uniform(size=self.N_NEURONS, low=200, high=300)
        self.threshold = rng.uniform(size=self.N_NEURONS, low=-1.0, high=1.0)
        self.basis_neurons.set_gain_bias(self.max_rates, self.threshold)
        self.basis_neurons.bias_signal.value[:] = self.basis_neurons.bias
        
        # 3: create encoder/decoder circuit
        self.enc_basis = Encoder(self.basis_signal, self.basis_neurons)
        self.enc_direc = Encoder(self.direction_signal, self.basis_neurons)
        
        self.enc_basis.weights *= self.basis_neurons.gain[:, None]
        self.enc_direc.weights *= self.basis_neurons.gain[:, None]
        
        # 4: train encoder/decoder circuit
        
        # 4.1: Sample a possible trajectory
        self.data_directions = (rng.randn(self.clamp_steps, self.N_DIMENSIONS))
        self.data_directions = scipy.signal.lfilter(np.ones(64)/256.0, [1], self.data_directions)
        self.data_positions_real = np.cumsum(self.data_directions, axis=0)
        self.data_positions_table = np.zeros((len(self.data_positions_real), self.N_POSITIONS, self.N_DIMENSIONS))
        for pos, table in zip(self.data_positions_real, self.data_positions_table):
            pos_x = pos[0]
            pos_y = pos[1]
            pos_x = pos_x % self.N_POSITIONS  # -- ensures non-neg, but still float
            pos_y = pos_y % self.N_POSITIONS  # -- ensures non-neg, but still float
            posi_x = int(pos_x)
            posi_y = int(pos_y)
            posf_x = pos_x - posi_x
            posf_y = pos_y - posi_y
            table[posi_x, posi_y ] = ( 2 - posf_x - posf_y ) / 2
            table[(posi_x + 1) % self.N_POSITIONS, (posi_y + 1) % self.N_POSITIONS] = ( posf_x + posf_y ) / 2
            table[posi_x, (posi_y + 1) % self.N_POSITIONS] = ( 1 - posf_x + posf_y ) / 2
            table[(posi_x + 1) % self.N_POSITIONS, posi_y] = ( posf_x + 1 - posf_y ) / 2
        
        # 4.3: Convert data positions to Fourier domain
        #blobs = scipy.signal.lfilter([.15, .7, .15], [1], data_positions_table)
        self.basis_blobs = scipy.fftpack.fft(self.data_positions_table)
        self.basis_blobs[:, self.N_BASIS:] = 0
        self.data_basis = np.empty((len(self.basis_blobs), 2 * self.N_BASIS, self.N_DIMENSIONS))
        self.data_basis[:, ::2] = np.real(self.basis_blobs[:, :self.N_BASIS])
        self.data_basis[:, 1::2] = np.imag(self.basis_blobs[:, :self.N_BASIS])
        self.data_basis_next = np.zeros_like(self.data_basis)
        self.data_basis_next[:-1] = self.data_basis[1:]
        self.data_basis_next[-1] = self.data_basis[0]
        
        self.d2 = scipy.fftpack.ifft(self.basis_blobs)
        
        self.direction_coef = 1.0 / np.max(abs(self.data_directions))

        self.clamp_basis = Clamp(np.asarray(self.data_basis), dt=self.clamp_dt)
        self.clamp_direc = Clamp(np.asarray(self.data_directions)[:, None] * self.direction_coef,
                            dt=self.clamp_dt)
        self.clamp_basis_out = Clamp(np.asarray(self.data_basis_next), dt=self.clamp_dt)
        self.clamp_positions = Clamp(np.asarray(self.data_positions_table), dt=self.clamp_dt)
        print(self.data_positions_table.shape)
        # 4.4: Run simulator
        training_model = SimModel()
        #nl_readout = training_model.signal(N_NEURONS)
        training_model.signals.update(
            [self.basis_signal, self.position_signal, self.direction_signal,
             self.basis_neurons.input_signal, self.basis_neurons.bias_signal,
             self.basis_neurons.output_signal,
             self.clamp_basis.output_signal,
             self.clamp_direc.output_signal,
             self.clamp_basis_out.output_signal,
             self.clamp_positions.output_signal,
            ])
        training_model.nonlinearities.update([self.basis_neurons,
                                              self.clamp_basis, self.clamp_direc,
                                              self.clamp_basis_out, self.clamp_positions])
        training_model.encoders.update([self.enc_basis, self.enc_direc])
        training_model.decoder(self.clamp_basis, self.basis_signal, 1.0)
        training_model.decoder(self.clamp_direc, self.direction_signal, 1.0)
        training_model.transform(1.0, self.direction_signal, self.direction_signal)
        #training_model.decoder(basis_neurons, nl_readout, 1.0)
        #training_model.transform(1.0, nl_readout, nl_readout)
        
        simple_filter_decoded_signal(training_model, self.basis_signal, pstc=.01)
        simple_filter_decoded_signal(training_model, self.position_signal, pstc=.01)
        
        p_clamp_direc = probe_nl(training_model, self.clamp_direc)
        p_clamp_basis = probe_nl(training_model, self.clamp_basis)
        p_basis_neurons = probe_nl(training_model, self.basis_neurons)
        p_basis_signal = training_model.probe(self.basis_signal, dt=0)
        p_clamp_basis_out = probe_nl(training_model, self.clamp_basis_out)
        p_clamp_positions = probe_nl(training_model, self.clamp_positions)
        
        sim = Simulator(training_model)
        sim.run_steps(self.clamp_steps)
        
        A = np.asarray(sim.probe_data(p_basis_neurons))
        d_clamp_basis_out = sim.probe_data(p_clamp_basis_out)
        d_clamp_positions = sim.probe_data(p_clamp_positions)
        
        basis_readout_weights, res5, rank5, s5 = np.linalg.lstsq(A,
                                                                 d_clamp_basis_out,
                                                                 rcond=0.01)
        posit_readout_weights, res6, rank6, s6 = np.linalg.lstsq(A,
                                                                 d_clamp_positions,
                                                                 rcond=0.01)
        print( self.basis_neurons )
        print( self.position_signal )
        print(posit_readout_weights.shape)
        self.dec_basis = Decoder(self.basis_neurons, self.basis_signal,
                                 weights=basis_readout_weights.T)
        self.dec_posit = Decoder(self.basis_neurons, self.position_signal,
                                 weights=posit_readout_weights.T)
    
    def test_model( self, direction_data, injected_data=None, steps=1000, b_ratio=0.97 ):
        """ direction_data is a numpy array
            injected_data is a tuple of numpy arrays """
        
        data_injected_fourier_pos = np.zeros( ( self.clamp_steps, 2 * self.N_BASIS, self.N_DIMENSIONS ) )
        if injected_data != None:
            data_injected_fourier_pos = build_table_2d( injected_data )

        self.clamp_injected_fourier_pos = Clamp(np.asarray(data_injected_fourier_pos), dt=self.clamp_dt)
        
        self.testing_model = SimModel()
        self.testing_model.signals.update(
            [self.basis_signal, self.position_signal, self.direction_signal,
             self.basis_neurons.input_signal, self.basis_neurons.bias_signal,
             self.basis_neurons.output_signal, self.clamp_injected_fourier_pos.output_signal
            ])
        self.testing_model.nonlinearities.update([self.basis_neurons, self.clamp_injected_fourier_pos])
        self.testing_model.encoders.update([self.enc_basis, self.enc_direc])
        self.testing_model.decoders.update([self.dec_basis, self.dec_posit])
        
        simple_filter_decoded_signal(self.testing_model, self.position_signal, pstc=.005)
        
        # -- add direction clamp (using feedback for basis)
        self.clamp_direc.data = direction_data
        self.testing_model.signals.add(self.clamp_direc.output_signal)
        self.testing_model.nonlinearities.add(self.clamp_direc)
        self.testing_model.decoder(self.clamp_direc, self.direction_signal, 1.0)
        self.testing_model.decoder(self.clamp_injected_fourier_pos, self.clamp_injected_fourier_pos.output_signal, 1.0)
        
        self.testing_model.transform(1.0, self.direction_signal, self.direction_signal)
        a, b = filter_coefs(0.005, self.testing_model.dt)
        self.testing_model.filter(a, self.basis_signal, self.basis_signal)
        self.testing_model.transform(b*b_ratio, self.basis_signal, self.basis_signal)
        self.testing_model.transform(b*(1-b_ratio), self.clamp_injected_fourier_pos.output_signal, self.basis_signal)
        
        self.tp_basis = self.testing_model.probe(self.basis_signal, dt=0)
        self.tp_posit = self.testing_model.probe(self.position_signal, dt=0)
        self.tp_neurons = probe_nl(self.testing_model, self.basis_neurons)
        self.tp_dir = self.testing_model.probe(self.direction_signal, dt=0)
        
        self.test_sim = Simulator(self.testing_model)
        self.test_sim.run_steps( steps )
        
        self.calculated_position = get_position_2d( self.test_sim.probe_data(self.tp_posit).T, precise=True )
        self.theoretical_position = (np.cumsum(self.test_sim.probe_data(self.tp_dir).ravel()) *(32.0/256.0)) % self.N_POSITIONS
    
    def discrepancy( self, ratio, direction_data, injected_data, steps=1000 ):
        """ Runs the test model, using 'ratio' as the ratio of the 'b' value of the basis_signal and 
            clamp_injected_fourier_pos signal transform 
            This function returns a value representing how close the model results are to what the 'actual'
            result should be.
            This function is designed to be used by hyperopt """
        
        self.test_model( direction_data, injected_data, steps=steps, b_ratio=ratio )
        
        # Check at different offsets of theoretical position, since the start point is random
        min_val = float("inf")
        density = 2
        for i in xrange( self.N_POSITIONS * density ):
            # TODO: might need a third option of subtracting in the np.min() statement
            diff = np.minimum( np.abs( self.calculated_position - \
                                      ( self.theoretical_position + i / float( density ) ) % self.N_POSITIONS ),  
                               np.abs( self.calculated_position + self.N_POSITIONS - \
                                      ( self.theoretical_position + i / float( density ) ) % self.N_POSITIONS ) )
            val = np.sum( diff )
            if val < min_val:
                min_val = val
                
        return min_val

    def hyperopt_wrapper( self, ratio ):
        
        direction_data = np.ones( self.clamp_steps ) * 0.3
        
        # Test with no injected data
        injected_data = None
        
        return self.discrepancy( ratio, direction_data, injected_data, steps=1000 )
    
    def plot_results( self ):
        
        fig = plt.figure(figsize=(30,15), dpi=1000)
        plt.subplot(2, 4, 1)
        plt.plot(self.data_positions_real)
        plt.title('pos(t)')
        plt.subplot(2, 4, 2)
        plt.imshow(self.data_basis, aspect='auto')
        plt.title('fft(pos table)')
        plt.subplot(2, 4, 3)
        plt.imshow(np.real(self.d2).T, origin='lower', aspect='auto')
        plt.title('ifft(fft(pos table))')
    
        plt.subplot(2, 4, 5)
        plt.plot(self.test_sim.probe_data(self.tp_dir).ravel())
        plt.title('v(t)')
    
        plt.subplot(2, 4, 6)
        plt.imshow(self.test_sim.probe_data(self.tp_basis), aspect='auto')
        plt.title('basis readout')
    
        plt.subplot(2, 4, 7)
        plt.imshow(self.test_sim.probe_data(self.tp_posit).T, origin='lower', aspect='auto')
        plt.title('position readout')
        
        plt.subplot(2, 4, 8)
        plt.plot( self.calculated_position )
        plt.title('calculated position')
        
        plt.subplot(2, 4, 4)
        #plt.plot( (np.cumsum(self.test_sim.probe_data(self.tp_dir).ravel()) *(64.0/256.0)) % self.N_POSITIONS )
        plt.plot( self.theoretical_position )
        plt.title('theoretical position')
    
        plt.show()
        
    def plot_results_2d( self ):
        
        fig = plt.figure(figsize=(30,15), dpi=1000)
        plt.subplot(2, 4, 1)
        plt.plot(self.data_positions_real)
        plt.title('pos(t)')
        plt.subplot(2, 4, 2)
        plt.imshow(self.data_basis[:,:,0], aspect='auto')
        plt.title('fft(pos table)')
        plt.subplot(2, 4, 3)
        plt.imshow(np.real(self.d2)[:,:,0].T, origin='lower', aspect='auto')
        plt.title('ifft(fft(pos table))')
    
        plt.subplot(2, 4, 5)
        plt.plot(self.test_sim.probe_data(self.tp_dir).ravel())
        plt.title('v(t)')
    
        plt.subplot(2, 4, 6)
        plt.imshow(self.test_sim.probe_data(self.tp_basis), aspect='auto')
        plt.title('basis readout')
    
        plt.subplot(2, 4, 7)
        plt.imshow(self.test_sim.probe_data(self.tp_posit).T, origin='lower', aspect='auto')
        plt.title('position readout')
        
        plt.subplot(2, 4, 8)
        plt.plot( self.calculated_position )
        plt.title('calculated position')
        
        plt.subplot(2, 4, 4)
        #plt.plot( (np.cumsum(self.test_sim.probe_data(self.tp_dir).ravel()) *(64.0/256.0)) % self.N_POSITIONS )
        plt.plot( self.theoretical_position )
        plt.title('theoretical position')
    
        plt.show()
        
posecells = PosecellNetwork()

```
